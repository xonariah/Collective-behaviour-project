Graphs that show how the DoA and DoS changed

```python
import json
import matplotlib.pyplot as plt

with open('json/with_collison_penalty.json', 'r') as file:
    data = json.load(file)

doa_values = data.get('DOA_epi', [])
dos_values = data.get('DOS_epi', [])

doa_episodes, doa_values_list = zip(*doa_values) if doa_values else ([], [])
dos_episodes, dos_values_list = zip(*dos_values) if dos_values else ([], [])

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(dos_episodes, dos_values_list, label="DoS", color="blue")
plt.title("With collision penalty")
plt.xlabel("Episodes")
plt.ylabel("DoS")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(doa_episodes, doa_values_list, label="DoA", color="green")
plt.xlabel("Episodes")
plt.ylabel("DoA")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
with open('json\without_collision_penalty.json', 'r') as file:
    data = json.load(file)

doa_values = data.get('DOA_epi', [])
dos_values = data.get('DOS_epi', [])

doa_episodes, doa_values_list = zip(*doa_values) if doa_values else ([], [])
dos_episodes, dos_values_list = zip(*dos_values) if dos_values else ([], [])

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(dos_episodes, dos_values_list, label="DoS", color="blue")
plt.title("Without collision penalty")
plt.xlabel("Episodes")
plt.ylabel("DoS")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(doa_episodes, doa_values_list, label="DoA", color="green")
plt.xlabel("Episodes")
plt.ylabel("DoA")
plt.grid(True)

plt.tight_layout()
plt.show()
```
